```python
X=rand.rand(10,2)
plt.scatter(X[:,0],X[:,1],s=100)
```

```python
# calculate the distance between pair points
dist_sq=np.sum((X[:,np.newaxis,:]-X[np.newaxis,:,:])**2,axis=-1)
```

```python
# This give full sort which more than what we need.
nearest= np.argsort(dist_sq,axis=1) # by row
# We can use argpartition to get k nearest neighbors 
k=2
nearest_partition=np.argpartition(dist_sq,k+1,axis=1) # by row
nearest_partition
```

```python
plt.scatter(X[:,0],X[:,1],s=100)
k=2
for i in range(X.shape[0]):
    for j in nearest_partition[i,:k+1]:
        plt.plot(*zip(X[j],X[i]), color='black')
```

```python
nearest_partition[1,k+1]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
hourly= pd.read_csv('FremontHourly.csv',index_col='Date',parse_dates=True)
hourly.columns=['northbound','southbound']
hourly['total']=hourly['northbound'] + hourly['southbound']
```

```python
hourly.head()
```

```python
# resample data into daily and weekly
daily=hourly.resample('d','sum')
weekly=hourly.resample('w','sum')
```

```python
daily.head()
```

```python
weekly[['northbound','southbound','total']].plot()
plt.ylabel('weekly riders');
```

```python
# calculate the moving average. for each day take an average of 30 days period around it 
daily.rolling_mean(daily['total'], 30).plot();
```

```python
daily['total'].rolling(30,center=True).mean().plot();
```

```python
# Define a function which returns the hours of daylight
# given the day of the year, from 0 to 365

def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given date"""
    diff = date - pd.to_datetime('2000, 12, 21')
    day = diff.total_seconds() / 24. / 3600
    day %= 365.25
    m = 1. - np.tan(np.radians(latitude)) * np.tan(np.radians(axis) * np.cos(day * np.pi / 182.625))
    m = max(0, min(m, 2))
    return 24. * np.degrees(np.arccos(1 - m)) / 180.

# I use list function because map return an object. so I need to list the values. 
weekly['daylight'] = list(map(hours_of_daylight, weekly.index)) 
daily['daylight'] =list(map(hours_of_daylight, daily.index))
```

```python
weekly['daylight'].plot();
plt.ylabel('daylight hours in Seattle')
# notice that datelight hours are more than 8 hours in December and 16 hours in June
```

```python
plt.scatter(weekly['daylight'],weekly['total']);
plt.xlabel('daylight hours')
plt.ylabel('weekly bicycle traffic')
# this plot help us to say the pattern between two varialbes
```

```python
from sklearn.linear_model import LinearRegression

X=weekly[['daylight']].to_dense()
Y=weekly['total']
clf = LinearRegression(fit_intercept=True).fit(X, Y)

weekly['daylight_trend']=clf.predict(X)
weekly['daylight_corrected_trend']=weekly['total']-weekly['daylight_trend']+weekly['daylight_trend'].mean()

xfit=np.linspace(7,17)
yfit=clf.predict(xfit[:,None])

plt.scatter(weekly['daylight'],weekly['total'])
plt.plot(xfit,yfit,'-k')
plt.title('Bicyle traffic through the year')
plt.xlabel('daylight hours')
plt.ylabel('Bicyle traffic in weely');
```

```python
weekly.head()
```

```python
print(clf.coef_[0])
```

```python
trend=clf.predict(weekly[['daylight']].as_matrix())
plt.scatter(weekly['daylight'],weekly['total']-trend + np.mean(trend))
plt.plot(xfit,np.mean(trend)+0*yfit,'-k')
plt.title('weekly traffic detrended')
plt.xlabel('daylight hours')
plt.ylabel('adjusted weekly count');
```

```python
weekly[['total','daylight_trend']].plot()
plt.ylabel('total weekly riders');
```

```python
weekly['daylight_corrected_trend'].plot()
rms=np.std(weekly['daylight_corrected_trend'])
plt.ylabel('adjusted total weekly riders')
print('root mean square about trend:{0:.0f} riders'.format(rms))
```

## Accounting for day of week 

```python
days= ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
daily['dayofweek']= daily['total'].index.dayofweek

group= daily.groupby("dayofweek")['total'].mean()
group.index=days

group.plot()
plt.gca().set_xticklabels(group.index);
```

```python
for i in range(7):
    daily[days[i]]=(daily.index.dayofweek==i).astype(float)

X=daily[days+['daylight']]
y= daily['total']

clf=LinearRegression().fit(X,y)
daily['dayofweek_trend']=clf.predict(X)
daily[['total','dayofweek_trend']].plot();
```

```python
daily['dayofweek_corrected']=daily['total'] - daily['dayofweek_trend'] + daily['dayofweek_trend'].mean()
print('rms={0:.0f}'.format(np.std(daily['dayofweek_corrected'])))
daily['dayofweek_corrected'].plot();
```

## Accounting for temperature and preciption 

```python
weather=pd.read_csv('SeaTacWeather.csv', index_col='DATE',parse_dates=True, usecols=[2, 3, 6, 7])
# change the tempreture from C to F
weather['TMAX']=0.18* weather['TMAX']+32
weather['TMIN']=0.18* weather['TMIN']+32

# change prcipe from 1/10 mm to inches

weather['PRCP'] /=254

weather['TMAX'].resample('w','max').plot()
weather['TMIN'].resample('w','min').plot()
plt.title('Weekly temperature Extermes(F)')
plt.ylabel('Temprature Extremes in Seattle');
```

```python
weather['PRCP'].resample('w','sum').plot()
plt.ylabel('Weekly precipitation (inches)')
plt.title('precipitation in Seattle');
```

```python
# join weather and daily 
daily=daily.join(weather)
columns=days + ['daylight', 'TMIN', 'TMAX', 'PRCP']

X=daily[columns]
y=daily['total']

clf=LinearRegression().fit(X,y)
daily['overall_trend']=clf.predict(X)
daily[['total','overall_trend']].plot()
plt.ylabel('total daily riders');
```

```python
# detrend

daily['overall_trend_corrected']=daily['total']-daily['overall_trend']+daily['overall_trend'].mean()
print('rms {0:.0f}'.format(np.std(daily['overall_trend'])))
daily['overall_trend_corrected'].plot();
plt.ylabel('bicycle daily traffic');
```

```python
daily['overall_trend_corrected'].rolling(30,center=True).mean().plot();
```
